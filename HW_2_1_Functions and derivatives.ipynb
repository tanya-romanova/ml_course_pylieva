1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def plot_func_and_derivative(f, f_sym, df, df_sym, x_max):
  x = np.linspace(0, x_max, 100)

  # Створення графіку
  plt.figure(figsize=(10, 5))
  plt.plot(x, f, label=f"f(x) = {f_sym}", color='blue')
  plt.plot(x, df, label=f"f'(x) = {df_sym}", color='green')

  # Додавання легенди, міток та заголовка
  plt.title(f"Графік функції {f_sym} та її похідної")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.legend()
  plt.grid(True)

  plt.show()
```

```python
x_sym = sp.symbols('x')

f_sym = 10 * x_sym - 5
df_sym = sp.diff(f_sym, x_sym)
df_sym
```

```python
x_max = 10
n_pts = 100
x = np.linspace(0, x_max, n_pts)

f = 10 * x - 5
df = [10] * n_pts

plot_func_and_derivative(f, f_sym, df, df_sym, x_max)
```

Зміна кількості годин вивчення збільшує оцінку на величину, в 10 разів більшу за величину зміни. Це не залежить від того, скільки годин вчився студент.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x_sym = sp.symbols('x')

f_sym = sp.sin(x_sym)
df_sym = sp.diff(f_sym, x_sym)
df_sym
```

```python
x_max = 10
n_pts = 100
x = np.linspace(0, x_max, n_pts)

f = np.sin(x)
df = np.cos(x)

plot_func_and_derivative(f, f_sym, df, df_sym, x_max)
```

```python
x_sym = sp.symbols('x')

f_sym = sp.ln(x_sym)
df_sym = sp.diff(f_sym, x_sym)
df_sym
```

```python
x_max = 10
n_pts = 100
x = np.linspace(0, x_max, n_pts)

np.seterr(divide='ignore', invalid='ignore')
f = np.log(x)
df = 1/x

plot_func_and_derivative(f, f_sym, df, df_sym, x_max)
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x_sym = sp.symbols('x')

f_sym = x_sym**2 - 5 * x_sym + 1
df_sym = sp.diff(f_sym, x_sym)
df_sym
```

```python
x_max = 100
n_pts = 100
x = np.linspace(0, x_max, n_pts)

f = x**2 - 5 * x + 1
df = 2 * x - 5

plot_func_and_derivative(f, f_sym, df, df_sym, x_max)
```
